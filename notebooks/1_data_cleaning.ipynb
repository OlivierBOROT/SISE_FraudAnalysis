## IDAvisAutorisationCheque
->  identifiant de la transaction en cours

```python
# Non numerical value detection
df[df["IDAvisAutorisationCheque"].str.isalpha().fillna(False)]
```

```python
# Get rid of the non numerical values (line containing duplicated headers)
df_cleaned = df[~df["IDAvisAutorisationCheque"].str.isalpha().fillna(False)]
# Converting to string
df_cleaned["IDAvisAutorisationCheque"] = df_cleaned["IDAvisAutorisationCheque"].astype(
    str
)
```

## FlagImpaye
-> acception (0) ou refus de la transaction (1)

```python
df_cleaned["FlagImpaye"].unique()
```

```python
# transformation en flag numerique
df_cleaned["FlagImpaye"] = df_cleaned["FlagImpaye"].astype(int)
```

## Montant
-> montant de la transaction

```python
# checking for non numerical values
df_cleaned[df_cleaned["Montant"].str.isalpha().fillna(False)]
```

```python
# Replacing the "," character and Rounding to 2 decimals
df_cleaned["Montant"] = (
    df_cleaned["Montant"].str.replace(",", ".").astype(float).round(2)
)
```

## DateTransaction
-> date de la transaction

```python
df_cleaned["DateTransaction"] = pd.to_datetime(
    df_cleaned["DateTransaction"], format="%Y-%m-%d %H:%M:%S"
)
```

## Code decision
-> il s’agit d’une variable qui peut prendre ici 4 valeurs:  
- 0 : la transaction a été acceptée par le magasin
- 1 : la transaction et donc le client fait partie d’une liste blanche (bons payeurs).
Vous n’en rencontrerez pas dans cette base de données
- 2 : le client fait d’une partie d’une liste noire, son historique indique cet un
mauvais payer (des impayés en cours ou des incidents bancaires en cours),
sa transaction est alors automatiquement refusée
- 3 : client ayant était arrêté par le système par le passé pour une raison plus ou
moins fondée

```python
# Checking unique values
print("Na values in 'CodeDecision':", df_cleaned["CodeDecision"].isna().sum())

df_cleaned["CodeDecision"].value_counts()
```

```python
# Get rid of line containing Na Value and code 4 which is incorrectly flagged
df_cleaned = df_cleaned[~df_cleaned["CodeDecision"].isin([4])]
df_cleaned = df_cleaned.dropna(subset=["CodeDecision"])

# Transform the code decision into integers
df_cleaned["CodeDecision"] = df_cleaned["CodeDecision"].astype(int)
```

```python
# Rechecking
print("Na values in 'CodeDecision':", df_cleaned["CodeDecision"].isna().sum())

df_cleaned["CodeDecision"].value_counts()
```
